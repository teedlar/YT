```python
import requests
bnb = requests.get('https://api.binance.com/api/v1/ticker/price')
crypto = input('Enter Name : ')
print()
for show in bnb.json():
    if crypto.upper() in show['symbol']:
        print(show['symbol'],show['price'])
```

```python
import requests
bnb = requests.get('https://api.binance.com/api/v1/ticker/price')
crypto = input('Enter Name : ')
print()
for show in bnb.json():
    if crypto.upper().strip() in show['symbol']:
        print(show['symbol'],show['price'])
```

```python
import requests
bnb = requests.get('https://api.binance.com/api/v1/ticker/price')
crypto = input('Enter Name : ')
qty = 5
print()
for show in bnb.json():
    if crypto.upper().strip() in show['symbol']:
        print(show['symbol'],show['price'])
        print()
        print('Total Value '+str(float(show['price'])*qty))
```
