```python
import requests
```

## Wazirx Link

```python
data = requests.get('https://api.wazirx.com/api/v2/tickers')
```

## Json data

```python
print(data.json())
```

```python
print(data.json()['btcinr'])
```

```python
print(data.json()['btcinr']['last'])
```

```python
print(data.json()['xrpinr'])
```

```python
print(data.json()['xrpinr']['last'])
```
